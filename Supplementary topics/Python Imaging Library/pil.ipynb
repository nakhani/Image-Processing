```python
from PIL import Image
```

```python
image = Image.open('input/clouds-8459053_1280.jpg' )
image.show()
```

```python
from PIL import ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display

draw = ImageDraw.Draw(image)

text = "ای در دل من میل و تمنا همه تو\n\n وندر سـر من مایه سودا همه تو"
reshaped_text = arabic_reshaper.reshape(text) 
bidi_text = get_display(reshaped_text)

font = ImageFont.truetype("bnazanb.ttf", 80) 
text_width, text_height = draw.textsize(bidi_text, font=font)

image_width, image_height = image.size
x = (image_width - text_width) / 2
y = (image_height - text_height) / 2

draw.text((x, y), bidi_text , font=font, fill=	(19, 30, 58))

image.save('output/result.jpg')
image.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

r, g, b = image.split()


r_histogram = r.histogram()
g_histogram = g.histogram()
b_histogram = b.histogram()

r_histogram = np.array(r_histogram)
g_histogram = np.array(g_histogram)
b_histogram = np.array(b_histogram)


plt.plot(r_histogram, color='red', label='Red')
plt.plot(g_histogram, color='green', label='Green')
plt.plot(b_histogram, color='blue', label='Blue')

plt.title('RGB Histograms')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend()

plt.savefig('output/histogram_plot.png')
plt.tight_layout()
plt.show()
```

```python
from PIL import ImageOps

equalized_image = ImageOps.equalize(image)

equalized_image.save('output/equalized_image.png')
equalized_image.show()
```
